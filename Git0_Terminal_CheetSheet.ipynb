# Basic Terminal Navigation and Directory Commands
- **pwd**: Print Working Directory (shows the current directory).
  ```sh
  pwd
  ```

- **ls**: List files and directories.
  ```sh
  ls
  ```

- **ls -l**: List files and directories with detailed information.
  ```sh
  ls -l
  ```

- **cd [directory]**: Change Directory.
  ```sh
  cd practice/dir1
  ```

- **cd ..**: Go up one directory level.
  ```sh
  cd ..
  ```

- **mkdir [directory]**: Make a new directory.
  ```sh
  mkdir practice
  ```

### File Creation and Manipulation
- **touch [file]**: Create a new empty file.
  ```sh
  touch file1.txt
  ```

- **echo [text] > [file]**: Write text to a file (overwrites the file).
  ```sh
  echo "Hello, world!" > file1.txt
  ```

- **echo [text] >> [file]**: Append text to a file.
  ```sh
  echo "Appending text." >> file1.txt
  ```

- **cat [file]**: Display the contents of a file.
  ```sh
  cat file1.txt
  ```

### Copying and Moving Files
- **cp [source] [destination]**: Copy a file or directory.
  ```sh
  cp file1.txt dir2/
  ```

- **mv [source] [destination]**: Move (or rename) a file or directory.
  ```sh
  mv file1.txt dir1/
  ```

### Removing Files and Directories
- **rm [file]**: Remove a file.
  ```sh
  rm file2.txt
  ```

- **rm -r [directory]**: Remove a directory and its contents.
  ```sh
  rm -r dir3
  ```

### File and Directory Permissions
- **chmod [permissions] [file]**: Change the permissions of a file or directory.
  ```sh
  chmod 755 file1.txt
  ```

- **ls -l [file]**: Check the permissions of a file or directory.
  ```sh
  ls -l file1.txt
  ```

### Symbolic Links
- **ln -s [target] [link]**: Create a symbolic link to a file or directory.
  ```sh
  ln -s dir1/file1.txt dir2/file1_link.txt
  ```

### Additional Useful Commands
- **man [command]**: Display the manual page for a command (use `q` to quit).
  ```sh

  ```

- **clear**: Clear the terminal screen.
  ```sh
  clear
  ```

- **history**: Show the history of commands used in the terminal.

### Basic Usage of `find`

```bash
find /path/to/search -name "filename"
```

- `/path/to/search`: The directory where you want to start the search. You can use `.` to search in the current directory and its subdirectories.
- `-name "filename"`: The name of the file you are looking for. You can use wildcards (e.g., `*.txt` to find all text files).

### Examples

1. **Search for a specific file by name in the current directory and subdirectories:**
    ```bash
    find . -name "example.txt"
    ```

2. **Search for all `.txt` files in the `/home/user/documents` directory:**
    ```bash
    find /home/user/documents -name "*.txt"
    ```

3. **Search for files by name in multiple directories:**
    ```bash
    find /dir1 /dir2 /dir3 -name "example.txt"
    ```

4. **Search for files case-insensitively:**
    ```bash
    find . -iname "example.txt"
    ```

5. **Search for files with a specific extension:**
    ```bash
    find . -name "*.jpg"
    ```

### Advanced Usage

You can combine `find` with other options to perform more complex searches:

1. **Search for files modified within the last 7 days:**
    ```bash
    find . -name "*.txt" -mtime -7
    ```

2. **Search for empty files:**
    ```bash
    find . -type f -empty
    ```

3. **Search for files by size (e.g., files larger than 10MB):**
    ```bash
    find . -type f -size +10M
    ```

4. **Execute a command on each found file (e.g., list details):**
    ```bash
    find . -name "*.txt" -exec ls -l {} \;
    ```

Sure! Let's start from scratch to set up a Git repository in a directory on your desktop with files `a`, `b`, and `c`. We'll then go through a quick example to demonstrate how to use `git stash`.

### Step-by-Step Setup

1. **Create a Directory on Your Desktop**:

   Open your terminal and run the following commands to create a directory called `example-repo` on your desktop and navigate into it:

   ```bash
   cd ~/Desktop
   mkdir example-repo
   cd example-repo
   ```

2. **Initialize a Git Repository**:

   Initialize an empty Git repository in the `example-repo` directory:

   ```bash
   git init
   ```

3. **Create Files `a`, `b`, and `c`**:

   Create three text files (`a.txt`, `b.txt`, `c.txt`) in the repository:

   ```bash
   echo "This is file a" > a.txt
   echo "This is file b" > b.txt
   echo "This is file c" > c.txt
   ```

4. **Add and Commit the Files**:

   Add the files to the staging area and commit them:

   ```bash
   git add a.txt b.txt c.txt
   git commit -m "Initial commit with files a, b, and c"
   ```

5. **Create a New Branch**:

   Create a new branch called `b1`:

   ```bash
   git branch b1
   ```

6. **Switch to Branch `b1`**:

   Switch to the new branch `b1`:

   ```bash
   git checkout b1
   ```

### Example of Using `git stash`

1. **Make Changes in Branch `b1`**:

   Modify `a.txt`:

   ```bash
   echo "Some changes in branch b1" >> a.txt
   ```

2. **Stash the Changes**:

   Stash the changes to save them temporarily:

   ```bash
   git stash
   ```

3. **Switch Back to `main` Branch**:

   Switch back to the `main` branch:

   ```bash
   git checkout main
   ```

4. **Apply the Stash**:

   Switch back to `b1` and apply the stashed changes:

   ```bash
   git checkout b1
   git stash apply
   ```

